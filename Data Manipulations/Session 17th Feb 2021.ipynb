```python
import pandas as pd
import numpy as nm
import math
```

```python
data = pd.read_csv("E:/Data/HR Data.csv")
data.columns
```

```python
pivot = data.groupby(['Region','Gender']).agg({'Emp ID':'count','Salary':'mean','Weight in Kgs.':'mean'})
pivot

pivot1 = data.groupby(['Region','Gender']).agg({'Emp ID':'count', 'Salary':'mean'}).reset_index()
pivot1['Sum']=pivot1['Emp ID']+pivot1['Salary']
pivot1.drop([0,3], inplace=True)
pivot1
#data[['Region','Salary']]

```

```python
pd.merge(data, pivot1, how="left",left_on='Region', right_on='Region')
```

```python
pivot1
```

```python
pivot1_melted = pd.melt(pivot1,id_vars=['Region','Gender'], value_vars=['Emp ID','Salary','Sum'])
pivot1_melted
```

```python
pivot_unmelted = pivot1_melted.pivot(index=['Region','Gender'],columns='variable')
pivot_unmelted
```

```python
def factorial(num):
    if num<2: 
        return 1
    else:
        return num*factorial(num-1)
```

```python
factorial(15)
```
