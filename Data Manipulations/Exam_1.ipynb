### Even odd Function

```python
def even_odd(num):
    if num == 0:
        return "its zero"
    elif num%2 ==0:
        return "even"
    else:
        return "odd"
    
```

```python
even_odd(0)
```

### Factorial Function

```python
def factorial(num):
    fact = 1
    for i in range(1,num+1,1):
        fact = fact*i
    return fact
```

```python
factorial(5)
```

```python
def factorial_w(num):
    fact = 1
    while num > 0:
        fact = fact*num
        num = num-1
    return fact
```

```python
factorial_w(4)
```

### Traingle of *

```python
def trian(num):
    st = 1
    while st <= num:
        k = int((num - st)/2)
        print(" "*k, end = "")
        print("*"*st)
        st = st+2
    
```

```python
trian(5)
```
